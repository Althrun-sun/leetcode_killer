```python
import requests
from bs4 import BeautifulSoup
problem_data =raw_data
def get_difficulty(title: str) -> str:
    url = f"https://leetcode.com/problems/{title.replace(' ', '-').lower()}/"
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        difficulty = soup.find("div", class_="css-10o4wqw")
        if difficulty:
            return difficulty.text
        else:
            return "unknown"
    else:
        return "unknown"
    
problems_with_details = {}
for category, problem_list in problem_data.items():
    problems_with_details[category] = []
    for title in problem_list:
        problem_info = {
            "title": title,
            "url": f"https://leetcode.com/problems/{title.replace(' ', '-').lower()}/",
            "difficulty": get_difficulty(title),
            "completed": False
        }
        problems_with_details[category].append(problem_info)

print(problems_with_details)
```
